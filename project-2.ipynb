### who are the directors their movies get highest votes ?

```python

df_directors_votes = df.groupby('director').mean()['vote_average'].sort_values(ascending=False)[:10]
df_directors_votes
```

```python
# a function to create bar plot with parameters:
def create_plot(x,y,
                plot_color,
                xlab, ylab,
                lab_size,
                title,
                title_size,
                lab_and_title_color,
                xtick_rotation,
                tick_size,tick_color,fig_size_x=15,fig_size_y=10,plt_kind='bar'):
    fig = plt.figure(figsize = (fig_size_x, fig_size_y)) 
      # creating the bar plot 
    if plt_kind=='bar':
        plt.bar(x,y, color =plot_color);
    plt.xlabel(xlab,fontsize = lab_size,color = lab_and_title_color) 
    plt.ylabel(ylab,fontsize = lab_size,color = lab_and_title_color) 
    plt.title(title,fontsize=title_size,color = lab_and_title_color) 
    plt.xticks(rotation = xtick_rotation)
    plt.tick_params(labelsize=tick_size,labelcolor=tick_color)
```

```python
create_plot(x=df_directors_votes.index,y=df_directors_votes.values,
                plot_color='#42c8f5',
                xlab ="Director", ylab = "Votes Average", 
                lab_size=20,
                title="The Directors With The Highst Votes Average",
                title_size =20,
                lab_and_title_color='red',
                xtick_rotation =70,
                tick_size=14,
                tick_color='red')
```

### what is the optimal run-time for movie?

### I think that we need to explore two things to answer on this question:
1. on the aspect of single movie - is there is more profit or less for single movie.
2. on the overall aspect - is there are today more movies per year than in the past ?

### before we answer questions about profit, budget and revenue, we will remove all the columns with 0 in this columns.

```python
df.head()
```

```python
df_without_zero= df[(df["revenue_adj"]!=0) & (df["budget_adj"]!=0)]
df_without_zero.shape
```

```python
df_years= df_without_zero.groupby("release_year").mean()['profit_adj']
```

```python
plt.figure(figsize=(12, 10))
plt.xlabel("Realese Year",fontsize=20,color="green")
plt.ylabel("Average Profit for movie",fontsize=20,color="green")
plt.title("Changes in profit for single movie over the years",fontsize=20,color="green") 
sns.set_context("poster")
sns.lineplot(data=df_years,color='red');
```

###### we see that the average profit per movies declines over the years !

#### the mean profit for single movie, is less today than in the past !

#### now we go to the second aspect - movies per year:

```python

movies_per_year = df["release_year"].value_counts()
```

```python
plt.figure(figsize=(12, 10))
plt.xlabel("Realese Year",fontsize=20,color="green")
plt.ylabel("Movies counts",fontsize=20,color="green")
plt.title("Year Vs Movies counts",fontsize=20,color="green") 
sns.set_context("poster")
sns.lineplot(data=movies_per_year,color='blue');
```

#### we see very clearly that there are more movies per year over the years

 
#### on one hand there are more movies per year, on the other hand every single movie is less profitable.     
#### so let's check the sum of profit to every year !



```python
### we will use the df_without_zero - df we reduced the rows with 0 values 
### from the budjet and profit.
df_sum_profits = df_without_zero.groupby("release_year").sum()["profit_adj"]
# df_sum_profits
```

```python
plt.figure(figsize=(12, 10))
plt.xlabel("Realese Year",fontsize=20,color="green")
plt.ylabel("Sum profits per yesr",fontsize=20,color="green")
plt.title("profits change over years",fontsize=20,color="green") 
sns.set_context("poster")
sns.lineplot(data=df_sum_profits,color='orange');
```

#### because we saw that the number of movies per year rises over the years , and on the other hand the mean revenue from movie decline over the years , we want to se the sum of revenues per year to check if the global sum of the revenues from movies declines or rises. 

## conclusion:
### even the profit for a single value decline over the years, the sum of the profit has rose, because there are much more movies every year !!!
